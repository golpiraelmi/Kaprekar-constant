# Kaprekars's Constant 6174

Take any four-digit number, using at least two different digits. (Leading zeros are allowed.)
Arrange the digits in descending and then in ascending order to get two four-digit numbers, adding leading zeros if necessary.

Subtract the smaller number from the bigger number.
Go back to step 2 and repeat.

The above process, known as Kaprekar's routine, will always reach its fixed point, 6174, in at most 7 iterations. Once 6174 is reached, the process will continue yielding 7641 â€“ 1467 = 6174.

```python
import numpy as np
```

### Creating list of 4-digit numbers containing no repeating digits

```python
numbers_str=[]
for a in range(1,10):
    for b in range(0,10):
        if a!=b:
            for c in range(0,10):
                if b!=c and a!=c:
                    for d in range(0,10):
                        if c!=d and d!=b and d!=a:
                            numbers_str.append(str(a)+str(b)+str(c)+str(d))
```

### Sorting digits ascending and descending and calculating the difference 

```python
def Kap(numbers_str):
    numbers_asc=[] 
    numbers_dsc=[]

    for number in numbers_str:
    
        ascending = "".join(sorted(str(number)))
        numbers_asc.append(ascending)
    
        descending = "".join(sorted(str(number), reverse=True))
        numbers_dsc.append(descending)

    numbers_asc_int = [int(i) for i in numbers_asc]  # Converting strings to integers using a list comprehension
    numbers_dsc_int = [int(i) for i in numbers_dsc]  # Converting strings to integers using a list comprehension
    
    return list(np.array(numbers_dsc_int) - np.array(numbers_asc_int)) # difference between numbers in two lists
```

```python
Kap(Kap(Kap(Kap(Kap(Kap(Kap(numbers_str)))))))
```
